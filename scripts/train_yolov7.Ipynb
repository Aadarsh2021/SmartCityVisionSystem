```python
import cv2
import os

# Path to your training images
train_images_path = "D:/smart_city_system/data/bdd100k_augmented/images/train"

# Define target dimensions
TARGET_SIZE = (640, 640)

# Create a function to resize and pad images
def resize_and_pad(image):
    h, w = image.shape[:2]
    scale = min(TARGET_SIZE[0] / w, TARGET_SIZE[1] / h)
    new_w, new_h = int(w * scale), int(h * scale)
    resized = cv2.resize(image, (new_w, new_h), interpolation=cv2.INTER_AREA)
    
    # Create a black background
    padded = cv2.copyMakeBorder(
        resized,
        (TARGET_SIZE[1] - new_h) // 2,
        (TARGET_SIZE[1] - new_h + 1) // 2,
        (TARGET_SIZE[0] - new_w) // 2,
        (TARGET_SIZE[0] - new_w + 1) // 2,
        cv2.BORDER_CONSTANT,
        value=[0, 0, 0]  # Black padding
    )
    
    return padded

# Iterate over images and fix dimensions
for img_file in os.listdir(train_images_path):
    img_path = os.path.join(train_images_path, img_file)
    img = cv2.imread(img_path)
    
    if img is None:
        print(f"Corrupt or unreadable image: {img_file}")
        continue
    
    h, w = img.shape[:2]
    if (h, w) != TARGET_SIZE:
        print(f"Resizing: {img_file} from {h}x{w} to {TARGET_SIZE}")
        img_resized = resize_and_pad(img)
        cv2.imwrite(img_path, img_resized)

print("Resizing complete. All images are now 640x640.")
```
